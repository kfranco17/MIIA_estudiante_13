**Misión 2**

Reportar las principales características en los datos.
* *Versión base*: utilizar métricas básicas como conteos o promedios por categorías en los datos.
* *Milla extra*: utilizar visualizaciones simples como diagramas de barras, de pastel u otros.

```python
# código

    
# importación de paquetes necesarios
import matplotlib.pyplot as plot


# función que calcula la mediana de una lista de datos numéricos
def mediana(lista):
    datos = sorted(lista)
    index = len(datos) // 2
    if len(lista) % 2 != 0:
        return datos[index]
    return (datos[index - 1] + datos[index]) / 2
    
# función que calcula la desviación de una lista de datos numéricos
def desviacion_std(lista):
    n = len(lista)
    promedio = sum(lista)/n
    sum_cuadrado = sum((x-promedio)**2 for x in lista)
    return (sum_cuadrado/n)**0.5

# función que calcula la moda de una lista de datos
def moda(lista):
    frecuencia = {}
    for valor in lista:
        frecuencia[valor] = frecuencia.get(valor, 0) + 1
    mayor_frecuencia = max(frecuencia.values())
    moda = [llave for llave, valor in frecuencia.items()
                      if valor == mayor_frecuencia]
    return moda

# función que contabiliza elementos repetidos en lista
def contarElementosLista(lista):
        return {i:lista.count(i) for i in lista}

# función para determinar el nombre de la lista
def nombre_lista(lista):
    if lista[0] in lista_edad:
        nombre = 'edad'
    elif lista[0] in lista_escolaridad:
        nombre = 'escolaridad'
    elif lista[0] in lista_estado_civil:
        nombre = 'estado_civil'
    elif lista[0] in lista_estrato:
        nombre = 'estrato'
    elif lista[0] in lista_genero:
        nombre = 'genero'
    elif lista[0] in lista_promedio:
        nombre = 'promedio'
    else:
        nombre = 'region'
    return nombre   

#función que visualiza las características de una población dada

#funcion diagrama torta para datos categoricos
def diagrama_torta(lista):
    dict_datos = contarElementosLista(lista)
    nombre = nombre_lista(lista)
    llaves = list(dict_datos.keys())
    valor_dict = list(map(int, dict_datos.values()))
    
    import matplotlib.pyplot as plt
    etiquetas = llaves
    valores = valor_dict
    plt.pie(x=valores, labels=etiquetas, autopct='%1.2f%%')
    plt.title(nombre)
    leyenda = []
    for etiquetas, valores in zip(etiquetas, valores):
        leyenda.append(etiquetas + ' (' + str(valores)+')')
    plt.legend(leyenda,bbox_to_anchor=(1, 1))
    return plt.show()

#funcion histograma para datos numericos

def histograma(lista):
    nombre = nombre_lista(lista)       
    import matplotlib.pyplot as plt
    plt.hist(x=lista, color='#F2AB6D', rwidth=0.85,)
    plt.title('Histograma de '+ nombre)
    plt.xlabel(nombre)
    plt.ylabel('Frecuencia')
    return plt.show()

#funcion summary
def summary(lista):
    nombre = nombre_lista(lista)
    if type(lista[0]) == str:
        mod = moda(lista)
        total_datos = len(lista)
        conteo = contarElementosLista(lista)
        print('Metricas de: ' + str(nombre)+ '\nTotal de datos: '+ str(total_datos) +'\nDatos: '+ str(conteo)+ '\nLa moda es: '+ str(mod))
        diagrama_torta(lista)
    else:
        promedio_1 = sum(lista)/len(lista)
        promedio = round(promedio_1, 2)
        median = mediana(lista)
        ds_1 = desviacion_std(lista)
        ds = round(ds_1, 2)
        mod = moda(lista)
        maximo_val = max(lista)
        minimo_val = min(lista)
        total_datos = len(lista)
        print('Metricas de: '+ str(nombre) + '\nPromedio: '+ str(promedio) + '\nMediana: ' + str(median) + '\nDesviacion std: ' + str(ds) + '\nModa: ' + str(mod) + '\nMaximo: '+ str(maximo_val) + '\nMinimo: '+ str(minimo_val)+ '\nTotal de datos: '+ str(total_datos))
        histograma(lista)

        
for categoria in lista_personas:
    summary(categoria)  
#'''
#(nota que la variable 'poblacion' es local; podemos pasar cualquier población a la función)
#'''
```

```python
# código
base_datos = {}
for i in range(0,len(region)):
    base_datos[i] = [edad[i], escolaridad[i], estado_civil[i], estrato[i], genero[i], promedio[i], region[i]]
    
def ordenar_por_promedio_region(nombre_region):
    str_nombre_region=str(nombre_region)
    personas_region = []
    for x in range(0,len(region)):
        if str_nombre_region == base_datos[x][6]:
            lista_personas=list(base_datos.keys())
            personas_region.append(lista_personas[x])
    
    prom_persona = []
    for personas in personas_region:
        prom_persona.append(base_datos[personas][5])
        
    diccionario_posicion_promedio = dict(zip(personas_region,prom_persona))
    diccionario_promedio_ordenado = {}
    llaves_orden = sorted(diccionario_posicion_promedio , key=diccionario_posicion_promedio.get,reverse = True)
    
    for w in llaves_orden:
        diccionario_promedio_ordenado[w] = diccionario_posicion_promedio [w]
        region_orden_promedio= list(diccionario_promedio_ordenado.keys())
    
    genero_region=[]
    for y in region_orden_promedio:
        genero_region.append(base_datos[y][4])
    
    dicccionario_genero_promedio = dict(zip(region_orden_promedio,genero_region))
    
    personas_femenino=[]
    personas_masculino=[]
    personas_no_binario=[]
    personas_otro=[]
    for z in region_orden_promedio:
        if 'femenino' == dicccionario_genero_promedio[z]:
            lista_personas_genero=list(dicccionario_genero_promedio.keys())
            indice_lista_genero = lista_personas_genero.index(z)
            personas_femenino.append(lista_personas_genero[indice_lista_genero])
        if 'masculino' == dicccionario_genero_promedio[z]:
            lista_personas_genero=list(dicccionario_genero_promedio.keys())
            indice_lista_genero = lista_personas_genero.index(z)
            personas_masculino.append(lista_personas_genero[indice_lista_genero])
        if 'no binario' == dicccionario_genero_promedio[z]:
            lista_personas_genero=list(dicccionario_genero_promedio.keys())
            indice_lista_genero = lista_personas_genero.index(z)
            personas_no_binario.append(lista_personas_genero[indice_lista_genero])
        if 'otro' == dicccionario_genero_promedio[z]:
            lista_personas_genero=list(dicccionario_genero_promedio.keys())
            indice_lista_genero = lista_personas_genero.index(z)
            personas_otro.append(lista_personas_genero[indice_lista_genero])
    
    lista_genero_region = (personas_femenino, personas_masculino, personas_no_binario,personas_otro)
    generos_region = ('femenino', 'masculino', 'no binario', 'otro')
    diccionario_posicion_promedio_genero = dict(zip(generos_region,lista_genero_region))

    return diccionario_posicion_promedio_genero

#posiciones ordenadas por promedio y Region
orden_Andina = ordenar_por_promedio_region('Andina')
orden_Pacifica = ordenar_por_promedio_region('Pacifica')
orden_Orinoquia = ordenar_por_promedio_region('Orinoquia')
orden_Caribe = ordenar_por_promedio_region('Caribe')
orden_Amazonia = ordenar_por_promedio_region('Amazonia')


#Determinacion de orden de becas y becados
becados_region_caribe=[]
for i in range(len(orden_Caribe['otro'])):
    becados_region_caribe_1= [orden_Caribe['femenino'][i], orden_Caribe['masculino'][i], orden_Caribe['no binario'][i], orden_Caribe['otro'][i]]
    becados_region_caribe.extend(becados_region_caribe_1)
for j in range(len(orden_Caribe['otro']), len(orden_Caribe['no binario'])):
    becados_region_caribe_2= [orden_Caribe['femenino'][j], orden_Caribe['masculino'][j], orden_Caribe['no binario'][j]]
    becados_region_caribe.extend(becados_region_caribe_2)
for k in range(len(orden_Caribe['no binario']), len(orden_Caribe['masculino'])):
    becados_region_caribe_3= [orden_Caribe['femenino'][k], orden_Caribe['masculino'][k]]
    becados_region_caribe.extend(becados_region_caribe_3) 
for l in range(len(orden_Caribe['masculino']), len(orden_Caribe['femenino'])):
    becados_region_caribe_4= [orden_Caribe['femenino'][l]]
    becados_region_caribe.extend(becados_region_caribe_4)

becados_region_andina=[]
for i in range(len(orden_Andina['otro'])):
    becados_region_andina_1= [orden_Andina['masculino'][i], orden_Andina['femenino'][i], orden_Andina['no binario'][i], orden_Andina['otro'][i]]
    becados_region_andina.extend(becados_region_andina_1)
for j in range(len(orden_Andina['otro']), len(orden_Andina['no binario'])):
    becados_region_andina_2= [orden_Andina['masculino'][j], orden_Andina['femenino'][j], orden_Andina['no binario'][j]]
    becados_region_andina.extend(becados_region_andina_2)
for k in range(len(orden_Andina['no binario']), len(orden_Andina['femenino'])):
    becados_region_andina_3= [orden_Andina['masculino'][k], orden_Andina['femenino'][k]]
    becados_region_andina.extend(becados_region_andina_3) 
for l in range(len(orden_Andina['femenino']), len(orden_Andina['masculino'])):
    becados_region_andina_4= [orden_Andina['masculino'][l]]
    becados_region_andina.extend(becados_region_andina_4)

becados_region_Pacifica=[]
for j in range(len(orden_Pacifica['no binario'])):
    becados_region_Pacifica_2= [orden_Pacifica['femenino'][j], orden_Pacifica['masculino'][j], orden_Pacifica['no binario'][j]]
    becados_region_Pacifica.extend(becados_region_Pacifica_2)
for k in range(len(orden_Pacifica['no binario']), len(orden_Pacifica['masculino'])):
    becados_region_Pacifica_3= [orden_Pacifica['femenino'][k], orden_Pacifica['masculino'][k]]
    becados_region_Pacifica.extend(becados_region_Pacifica_3) 
for l in range(len(orden_Pacifica['masculino']), len(orden_Pacifica['femenino'])):
    becados_region_Pacifica_4= [orden_Pacifica['femenino'][l]]
    becados_region_Pacifica.extend(becados_region_Pacifica_4)
    
becados_region_Orinoquia=[]
for i in range(len(orden_Orinoquia['otro'])):
    becados_region_Orinoquia_1= [orden_Orinoquia['femenino'][i], orden_Orinoquia['masculino'][i], orden_Orinoquia['otro'][i]]
    becados_region_Orinoquia.extend(becados_region_Orinoquia_1)
for k in range(len(orden_Orinoquia['otro']), len(orden_Orinoquia['masculino'])):
    becados_region_Orinoquia_3= [orden_Orinoquia['femenino'][k], orden_Orinoquia['masculino'][k]]
    becados_region_Orinoquia.extend(becados_region_Orinoquia_3) 

    becados_region_Amazonia=[]
for i in range(len(orden_Amazonia['otro'])):
    becados_region_Amazonia_1= [orden_Amazonia['masculino'][i], orden_Amazonia['femenino'][i], orden_Amazonia['no binario'][i], orden_Amazonia['otro'][i]]
    becados_region_Amazonia.extend(becados_region_Amazonia_1)
for k in range(len(orden_Amazonia['otro']), len(orden_Amazonia['femenino'])):
    becados_region_Amazonia_3= [orden_Amazonia['masculino'][k], orden_Amazonia['femenino'][k]]
    becados_region_Amazonia.extend(becados_region_Amazonia_3) 
for l in range(len(orden_Amazonia['femenino']), len(orden_Amazonia['masculino'])):
    becados_region_Amazonia_4= [orden_Amazonia['masculino'][l]]
    becados_region_Amazonia.extend(becados_region_Amazonia_4)

lista_becados= []
for i in range(len(becados_region_Amazonia)):
    becados_1 = [becados_region_caribe[i], becados_region_andina[i], becados_region_Pacifica[i], becados_region_Orinoquia[i], becados_region_Amazonia[i]]
    lista_becados.extend(becados_1)
for j in range(len(becados_region_Amazonia), len(becados_region_Orinoquia)):
    becados_2 = [becados_region_caribe[j], becados_region_andina[j], becados_region_Pacifica[j], becados_region_Orinoquia[j]]
    lista_becados.extend(becados_2)
for k in range(len(becados_region_Orinoquia), len(becados_region_Pacifica)):
    becados_3 = [becados_region_caribe[k], becados_region_andina[k], becados_region_Pacifica[k]]
    lista_becados.extend(becados_3)
for l in range(len(becados_region_Pacifica), len(becados_region_andina)):
    becados_4 = [becados_region_caribe[l], becados_region_andina[l]]
    lista_becados.extend(becados_4)
for m in range(len(becados_region_andina), len(becados_region_caribe)):
    becados_5 = [becados_region_caribe[m]]
    lista_becados.extend(becados_5)


#establecer numero de becas y resultado de becados

Numero_de_becas_region = 10
lista_becados_por_region = lista_becados[:Numero_de_becas_region]

print(lista_becados_por_region)
```
