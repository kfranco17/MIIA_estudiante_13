### Ejercicio 4
Utiliza `ipywidgets` para filtrar las observaciones de la base de datos de acuerdo con los siguentes criterios:
1. Nivel máximo de escolaridad alcanzado
2. Número de niños en el hogar
3. Número de adolescentes en el hogar

Asimismo, el 

```python
# YOUR CODE HERE

def valores_filtro_columna(columna):
    columna = columna.astype(str)       
    valores = columna.unique().tolist() 
    valores = sorted(valores)           
    valores.insert(0, "Todos")          
    return valores 


def filtrar(Educacion, hijos_casa, adolescentes_casa):
    # Guardamos las columnas de interés en un nuevo DataFrame.
    DF = mercadeo[['ID','Marital_Status','Income','Education', 'Kidhome', 'Teenhome']]
    # Los siguientes cuatro casos ajustan los registros a mostrar de acuerdo con la selección del usuario.
    if  Educacion != "Todos" and hijos_casa != "Todos":
        display(DF[(DF['Education'] == Educacion) & (DF["Kidhome"] == hijos_casa)
                   & (DF["Teenhome"] <= adolescentes_casa)])
    elif Educacion == "Todos":
        display(DF[(DF["Kidhome"] == hijos_casa) & (DF["Teenhome"] <= adolescentes_casa)])
    elif hijos_casa == "Todos":
        display(DF[(DF["Education"] == Educacion) & (DF["Kidhome"] <= hijos_casa)])
    else:
        display(DF[(DF["Teenhome"] <= adolescentes_casa)])
        
interactive_plot = interactive(
                                filtrar, 
                                Educacion = valores_filtro_columna(mercadeo["Education"]),
                                hijos_casa = (0, 5, 1),
                                adolescentes_casa = (0, 5, 1)
                              )
display(interactive_plot)
```

### Ejercicio 5
Utiliza `ipywidgets` y `plotly` para generar un gráfico con controles que:
1. Cuente con un control para seleccionar una de las variable de Gastos frente a los ingresos.
1. Genere un gráfico de dispersión
2. Genere un diagrama de caja.

```python
# YOUR CODE HERE
df = mercadeo[['Income','MntWines','MntFruits','MntMeatProducts','MntFishProducts','MntSweetProducts', 'MntGoldProds']]
df_unpvt = df.melt(id_vars=['Income'], var_name = 'tipo_gasto', value_name = 'gasto')


tipo = widgets.Dropdown(
                        description = 'Tipo de gasto:',
                        options = df_unpvt['tipo_gasto'].unique().tolist(),
                        style = {'description_width': 'initial'}       
                       )


fig1 = go.FigureWidget(
                        data = go.Scatter(name = 'Gasto - Ingresos', mode='markers'),  
    
                        layout = go.Layout(
                                            
                                            xaxis_title = 'Ingresos',
                                            yaxis_title = 'Gasto',
                                          )
                     )


fig2 = go.FigureWidget(
                        data = go.Box(name = 'Ingreso' ),  
    
                        layout = go.Layout(
                                            
                                            yaxis_title = 'Ingreso',
                                            
                                          )
                     )


def filtrar_list():
    filtrar_lista = [i for i in df_unpvt['tipo_gasto'] == tipo.value]
    df_temp = df_unpvt[filtrar_lista]
    fig1.data[0].x = df_temp['Income']
    fig1.data[0].y = df_temp['gasto'] 
    fig2.data[0].y = df_temp['Income'] 
    fig1.update_layout(title_text= 'Ingreso-'+ tipo.value)
    fig2.update_layout(title_text= 'Ingreso: '+ tipo.value)

def respuesta(change):
    filtrar()

tipo.observe(respuesta, names = 'value')  
filtrar_list()
parte_inf = widgets.HBox([tipo])   

visualizar = widgets.VBox(
              [parte_inf,
              fig1, fig2]
            )

display(visualizar)
```
